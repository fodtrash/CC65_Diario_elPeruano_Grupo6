```python
date_columns = ['FECHA_PUBLICACION', 'FECHA_CORTE']
min_year = 2013
max_year = 2025

print("--- Estandarización de Fechas a formato YYYYMMDD ---\n")

for col in date_columns:
    print(f"  Columna: {col}")

    # Convertir a string por si acaso quedaron como int o float
    consolidated_df[col] = consolidated_df[col].astype(str).str.strip()

    # Clasificar cada fila en uno de tres grupos
    mask_yyyymmdd  = consolidated_df[col].str.fullmatch(r'\d{8}')
    mask_ddmmyyyy  = consolidated_df[col].str.fullmatch(r'\d{2}/\d{2}/\d{4}')
    mask_otro      = ~mask_yyyymmdd & ~mask_ddmmyyyy

    print(f"  Formato YYYYMMDD  : {mask_yyyymmdd.sum():>7,} registros")
    print(f"  Formato DD/MM/YYYY: {mask_ddmmyyyy.sum():>7,} registros")
    print(f"  Formato desconocido:{mask_otro.sum():>6,} registros")

    # Verificar formato desconocido (si existe)
    if mask_otro.sum() > 0:
        print(f"\n  Ejemplos de formato desconocido:")
        print(consolidated_df.loc[mask_otro, col].head(10).to_string())

    # Convertir DD/MM/YYYY → YYYYMMDD
    if mask_ddmmyyyy.sum() > 0:
        convertidos = pd.to_datetime(
            consolidated_df.loc[mask_ddmmyyyy, col],
            format='%d/%m/%Y',
            errors='coerce'
        ).dt.strftime('%Y%m%d')

        # Verificar cuántos fallaron la conversión (deberían ser 0)
        fallidos = convertidos.isna().sum()
        if fallidos > 0:
            print(f"\n  {fallidos} registros no pudieron convertirse:")
            print(consolidated_df.loc[mask_ddmmyyyy & convertidos.isna(), col].head(5).to_string())

        consolidated_df.loc[mask_ddmmyyyy, col] = convertidos
        print(f"\n {mask_ddmmyyyy.sum() - fallidos:,} registros convertidos a YYYYMMDD.")

    # ── Estado final de la columna ───────────────────────────────
    final_valid = consolidated_df[col].str.fullmatch(r'\d{8}')
    print(f"\n  Estado final:")
    print(f"    En formato YYYYMMDD : {final_valid.sum():>7,}")
    print(f"    Fuera de formato    : {(~final_valid).sum():>7,}")

    # ── Verificación de rango de años ────────────────────────────
    fechas_dt = pd.to_datetime(consolidated_df.loc[final_valid, col], format='%Y%m%d', errors='coerce')
    fuera_rango = fechas_dt[(fechas_dt.dt.year < min_year) | (fechas_dt.dt.year > max_year)]

    if len(fuera_rango) > 0:
        print(f"\n  {len(fuera_rango):,} registros fuera del rango {min_year}-{max_year}:")
        print(fuera_rango.head(10).to_string())
    else:
        print(f"    Rango de años ({min_year}-{max_year})  : todos válidos")
```

```python
import unicodedata

# Mapa completo: variantes normales + strings corruptos con bytes incrustados
MAPA_DISPOSITIVO = {
    # Variantes normales
    'ORDENANZAS'              : 'ORDENANZA',
    'DECRETOS DE ALCALDIA'    : 'DECRETO DE ALCALDIA',
    'RES. ALCALDIA'           : 'RESOLUCION DE ALCALDIA',
    'FE DE ERRATAS'           : 'FE DE ERRATA',
    # Strings con bytes corruptos — mapeados directamente
    'DECRETOS DE ALCALD\xa1A' : 'DECRETO DE ALCALDIA',   # ¡ = \xa1
    'DECRETOS DE ALCALDA\xadA': 'DECRETO DE ALCALDIA',   # soft hyphen = \xad
    'RES. ALCALD\xa1A'        : 'RESOLUCION DE ALCALDIA', # ¡ = \xa1
    'RES. ALCALDA\xadA'       : 'RESOLUCION DE ALCALDIA', # soft hyphen = \xad
    'RESOLUCI\xa2N'           : 'RESOLUCION',             # ¢ = \xa2
    'RESOLUCIA\xb3N'          : 'RESOLUCION',             # ³ = \xb3
}

def normalizar_dispositivo(valor):
    if pd.isna(valor):
        return None
    valor = str(valor).strip()
    # Primero verificar si el string corrupto está en el mapa directo
    if valor in MAPA_DISPOSITIVO:
        return MAPA_DISPOSITIVO[valor]
    # Si no, aplicar normalización estándar (quitar tildes + mayúsculas)
    valor = unicodedata.normalize('NFD', valor)
    valor = ''.join(c for c in valor if unicodedata.category(c) != 'Mn')
    valor = re.sub(r'\s+', ' ', valor.upper().strip())
    return MAPA_DISPOSITIVO.get(valor, valor)

consolidated_df['DISPOSITIVO'] = consolidated_df['DISPOSITIVO'].apply(normalizar_dispositivo)

print(consolidated_df['DISPOSITIVO'].value_counts(dropna=False))
```

```python
def limpiar_numero(valor):
    if pd.isna(valor):
        return None
    valor = str(valor).strip()
    # Reparar prefijos corruptos de "N°"
    valor = valor.replace('NÂ°', 'No')
    valor = valor.replace('Nø', 'No')
    valor = valor.replace('Nª', 'No')
    valor = valor.replace('N.º', 'No')
    valor = valor.replace('N.°', 'No')
    valor = valor.replace('Mº ', 'No')
    valor = valor.replace('N°', 'No')
    valor = valor.replace('No º', 'No')
    valor = valor.replace('N¦', 'No')


    valor = re.sub(r'\s+', ' ', valor).strip()
    return valor

consolidated_df['NUMERO'] = consolidated_df['NUMERO'].apply(limpiar_numero)

# Verificar que no quedan caracteres no ASCII
mask = consolidated_df['NUMERO'].astype(str).str.contains(r'[^\x00-\x7F]', na=False)
print(f'Registros con caracteres no ASCII restantes: {mask.sum():,}')
if mask.sum() > 0:
    print('Ejemplos restantes:')
    for val in consolidated_df.loc[mask, 'NUMERO'].unique()[:100]:
        print(repr(val))
```

```python
# Análisis de distribuciones del corpus real

import numpy as np
from collections import Counter

N_REAL      = len(consolidated_df)
N_OBJETIVO  = 1_000_000
N_SINTETICO = N_OBJETIVO - N_REAL

print(f"Registros reales   : {N_REAL:>10,}")
print(f"Registros a generar: {N_SINTETICO:>10,}")
print(f"Registros objetivo : {N_OBJETIVO:>10,}")
print()

# Distribución de DISPOSITIVO
dist_dispositivo = consolidated_df["DISPOSITIVO"].value_counts(normalize=True)
print("Distribución DISPOSITIVO (top 10):")
print(dist_dispositivo.head(10).to_string())
print()

# Distribución de ENTIDAD (top 200 para mayor variedad)
dist_entidad = consolidated_df["ENTIDAD"].value_counts(normalize=True).head(200)
n_entidades  = consolidated_df["ENTIDAD"].nunique()
cobertura    = dist_entidad.sum() * 100
print(f"Entidades únicas totales : {n_entidades:,}")
print(f"Top-200 cubren el        : {cobertura:.1f}% del corpus")
print()

# Estadísticas de longitud de SUMILLA
lens_sumilla = consolidated_df["SUMILLA"].str.split().str.len()
print("Longitud SUMILLA (en palabras):")
print(f"  Minima  : {lens_sumilla.min()}")
print(f"  Media   : {lens_sumilla.mean():.1f}")
print(f"  Mediana : {lens_sumilla.median():.0f}")
print(f"  Maxima  : {lens_sumilla.max()}")

# Vocabulario de construcción de SUMILLA
# Extraemos unigramas frecuentes del corpus real para recombinar
todos_los_tokens = (
    consolidated_df["SUMILLA"]
    .str.upper()
    .str.findall(r"[A-Z\u00C0-\u00FF]{3,}")
    .explode()
)
vocab_freq = Counter(todos_los_tokens.dropna().tolist())
n_vocab = len(vocab_freq)
print(f"Vocabulario unico (palabras >=3 chars): {n_vocab:,}")
print("Top-20 tokens mas frecuentes:")
for tok, cnt in vocab_freq.most_common(20):
    print(f"  {tok:<30} {cnt:>8,}")
```

```python
# Generación de registros sintéticos
#
# Estrategia: Augmentación combinatoria guiada por distribuciones reales.
# Cada registro sintético se construye muestreando:
#   1. DISPOSITIVO        -> según distribución real (value_counts normalizados)
#   2. ENTIDAD            -> según distribución real (top-200)
#   3. FECHA_PUBLICACION / FECHA_CORTE -> fechas válidas dentro del rango 2013-2025
#   4. NUMERO             -> formato numérico basado en patrones del corpus real
#   5. SUMILLA            -> plantilla + vocabulario léxico extraído del corpus real

import random
from datetime import date, timedelta

random.seed(42)
np.random.seed(42)

pool_dispositivos  = dist_dispositivo.index.tolist()
probs_dispositivos = dist_dispositivo.values.tolist()

pool_entidades     = dist_entidad.index.tolist()
probs_entidades    = (dist_entidad / dist_entidad.sum()).values.tolist()

vocab_top = [tok for tok, _ in vocab_freq.most_common(2000)]

STOPWORDS = {"QUE", "DE", "LA", "EL", "LOS", "LAS", "DEL", "EN", "SE", "CON",
             "POR", "AL", "SU", "UNA", "UNO", "LO", "HA", "HAN", "PARA"}
vocab_content = [t for t in vocab_top if t not in STOPWORDS]

# Plantillas de SUMILLA extraídas del patrón léxico del Diario Oficial
PLANTILLAS_SUMILLA = [
    "{DISP} QUE {VERBO} {OBJ} DE {ENT}",
    "{DISP} No {NUM} - {ENT} - {OBJ}",
    "SE {VERBO} {OBJ} PARA LA {ENT}",
    "{DISP} QUE APRUEBA {OBJ} DEL {ENT}",
    "APRUEBAN {OBJ} DE LA {ENT}",
    "AUTORIZAN {OBJ} A LA {ENT}",
    "{DISP} QUE MODIFICA {OBJ} DE LA {ENT}",
    "DESIGNAN {OBJ} EN LA {ENT}",
    "DISPONEN {OBJ} RESPECTO A LA {ENT}",
    "{DISP} QUE ESTABLECE {OBJ} EN MATERIA DE {ENT}",
    "MODIFICAN {OBJ} DE LA {ENT}",
    "CREAN {OBJ} EN EL AMBITO DE LA {ENT}",
    "{DISP} QUE REGULA {OBJ} A CARGO DE LA {ENT}",
    "DEJAN SIN EFECTO {OBJ} DE LA {ENT}",
    "{DISP} QUE DEROGA {OBJ} EMITIDO POR LA {ENT}",
]

VERBOS = ["APRUEBA", "MODIFICA", "ESTABLECE", "REGULA", "AUTORIZA",
          "DISPONE", "DESIGNA", "CREA", "DEROGA", "SUSPENDE",
          "PRORROGA", "DECLARA", "DELEGA", "INCORPORA", "ASIGNA"]

# Funciones auxiliares

def fecha_aleatoria_str(anio_min=2013, anio_max=2025):
    inicio = date(anio_min, 1, 1)
    fin    = date(anio_max, 12, 31)
    delta  = (fin - inicio).days
    return (inicio + timedelta(days=random.randint(0, delta))).strftime("%Y%m%d")

def numero_sintetico(dispositivo):
    n    = random.randint(1, 9999)
    anio = random.randint(2013, 2025)
    if "DECRETO" in dispositivo:
        return f"No {n:04d}-{anio}"
    elif "RESOLUCION" in dispositivo:
        sufijo = random.choice(vocab_content[:50])[:4]
        return f"No {n:04d}-{anio}-{sufijo}"
    return f"No {n:04d}-{anio}"

def sumilla_sintetica(dispositivo, entidad):
    plantilla  = random.choice(PLANTILLAS_SUMILLA)
    obj_tokens = random.sample(vocab_content, k=random.randint(2, 4))
    obj_str    = " ".join(obj_tokens)
    verbo      = random.choice(VERBOS)
    num_str    = str(random.randint(1, 9999)).zfill(4)
    ent_short  = entidad[:40] if len(entidad) > 40 else entidad
    return (
        plantilla
        .replace("{DISP}",  dispositivo)
        .replace("{VERBO}", verbo)
        .replace("{OBJ}",   obj_str)
        .replace("{ENT}",   ent_short)
        .replace("{NUM}",   num_str)
    )

# Generación por lotes
LOTE = 50_000
sinteticos_acumulados = []
generados = 0

print(f"Generando {N_SINTETICO:,} registros sinteticos en lotes de {LOTE:,}...")

while generados < N_SINTETICO:
    n_lote = min(LOTE, N_SINTETICO - generados)

    dispositivos = np.random.choice(pool_dispositivos, size=n_lote, p=probs_dispositivos)
    entidades    = np.random.choice(pool_entidades,    size=n_lote, p=probs_entidades)
    fechas_pub   = [fecha_aleatoria_str() for _ in range(n_lote)]
    fechas_corte = [fecha_aleatoria_str() for _ in range(n_lote)]
    numeros      = [numero_sintetico(d)       for d in dispositivos]
    sumillas     = [sumilla_sintetica(d, e)   for d, e in zip(dispositivos, entidades)]

    lote_df = pd.DataFrame({
        "FECHA_PUBLICACION" : fechas_pub,
        "ENTIDAD"           : entidades,
        "DISPOSITIVO"       : dispositivos,
        "NUMERO"            : numeros,
        "SUMILLA"           : sumillas,
        "FECHA_CORTE"       : fechas_corte,
    })

    sinteticos_acumulados.append(lote_df)
    generados += n_lote
    print(f"  Lote completado: {generados:>10,} / {N_SINTETICO:,}")

df_sintetico = pd.concat(sinteticos_acumulados, ignore_index=True)
df_sintetico["ORIGEN"] = "SINTETICO"

n_sint = len(df_sintetico)
print(f"Registros sinteticos generados: {n_sint:,}")
```

```python
# Fusión dataset real + sintético

consolidated_df["ORIGEN"] = "REAL"

# Alinear columnas (consolidated_df ya no tiene OP/LINK tras el Bloque A)
COLS_FINALES = ["FECHA_PUBLICACION", "ENTIDAD", "DISPOSITIVO",
                "NUMERO", "SUMILLA", "FECHA_CORTE", "ORIGEN"]

dataset_final = (
    pd.concat([consolidated_df[COLS_FINALES], df_sintetico[COLS_FINALES]],
              ignore_index=True)
    .sample(frac=1, random_state=42)
    .reset_index(drop=True)
)

n_total  = len(dataset_final)
n_real   = (dataset_final["ORIGEN"] == "REAL").sum()
n_sint   = (dataset_final["ORIGEN"] == "SINTETICO").sum()
mem_mb   = dataset_final.memory_usage(deep=True).sum() / 1e6

print("=== Dataset final ensamblado ===")
print(f"Total registros    : {n_total:,}")
print(f"  Reales           : {n_real:,}")
print(f"  Sinteticos       : {n_sint:,}")
print(f"Columnas           : {list(dataset_final.columns)}")
print(f"Memoria            : {mem_mb:.1f} MB")
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle(
    "Dataset Final - Diario Oficial El Peruano\nVerificacion de Cumplimiento (1,000,000 registros)",
    fontsize=14, fontweight="bold", y=1.01
)

# Panel 1: Composición real vs sintético
ax1 = axes[0, 0]
origen_counts = dataset_final["ORIGEN"].value_counts()
colores = ["#2ecc71" if o == "REAL" else "#3498db" for o in origen_counts.index]
barras = ax1.bar(origen_counts.index, origen_counts.values, color=colores, edgecolor="white", width=0.5)
ax1.axhline(y=N_OBJETIVO, color="red", linestyle="--", alpha=0.7, label=f"Objetivo: {N_OBJETIVO:,}")
for bar, val in zip(barras, origen_counts.values):
    ax1.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5000,
             f"{val:,}", ha="center", va="bottom", fontsize=11, fontweight="bold")
ax1.set_title("Composicion del Dataset")
ax1.set_ylabel("Cantidad de registros")
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{int(x):,}"))
ax1.legend()
ax1.set_ylim(0, max(origen_counts.values) * 1.15)

# Panel 2: Distribución de DISPOSITIVO (top 10)
ax2 = axes[0, 1]
top_disp = dataset_final["DISPOSITIVO"].value_counts().head(10)
ax2.barh(top_disp.index[::-1], top_disp.values[::-1], color="#9b59b6", edgecolor="white")
ax2.set_title("Top 10 Tipos de Dispositivo")
ax2.set_xlabel("Registros")
ax2.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{int(x):,}"))
for i, (idx, val) in enumerate(zip(top_disp.index[::-1], top_disp.values[::-1])):
    ax2.text(val + 1000, i, f"{val:,}", va="center", fontsize=8)

# Panel 3: Registros por año (FECHA_PUBLICACION)
ax3 = axes[1, 0]
anios_real = pd.to_datetime(
    consolidated_df["FECHA_PUBLICACION"], format="%Y%m%d", errors="coerce"
).dt.year.value_counts().sort_index()
anios_sint = pd.to_datetime(
    df_sintetico["FECHA_PUBLICACION"], format="%Y%m%d", errors="coerce"
).dt.year.value_counts().sort_index()
ax3.bar(anios_real.index, anios_real.values, label="Real",      color="#2ecc71", alpha=0.8)
base_sint = anios_real.reindex(anios_sint.index, fill_value=0).values
ax3.bar(anios_sint.index, anios_sint.values, label="Sintetico", color="#3498db", alpha=0.8, bottom=base_sint)
ax3.set_title("Registros por Ano de Publicacion")
ax3.set_xlabel("Anio")
ax3.set_ylabel("Registros")
ax3.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{int(x):,}"))
ax3.legend()

# Panel 4: Resumen tabular de métricas clave
ax4 = axes[1, 1]
ax4.axis("off")

n_total   = len(dataset_final)
n_real_v  = (dataset_final["ORIGEN"] == "REAL").sum()
n_sint_v  = (dataset_final["ORIGEN"] == "SINTETICO").sum()
pct_real  = n_real_v / n_total * 100
pct_sint  = n_sint_v / n_total * 100
n_nulos   = dataset_final.isnull().sum().sum()
n_disp    = dataset_final["DISPOSITIVO"].nunique()
n_ent     = dataset_final["ENTIDAD"].nunique()
long_med  = dataset_final["SUMILLA"].str.split().str.len().mean()
objetivo  = "SI" if n_total >= N_OBJETIVO else "NO"

metricas = [
    ["Metrica",                     "Valor"],
    ["Total registros",             f"{n_total:,}"],
    ["  Reales",                    f"{n_real_v:,}  ({pct_real:.1f}%)"],
    ["  Sinteticos",                f"{n_sint_v:,}  ({pct_sint:.1f}%)"],
    ["Columnas",                    str(len(dataset_final.columns))],
    ["Nulos totales",               str(n_nulos)],
    ["Tipos de DISPOSITIVO",        str(n_disp)],
    ["Entidades unicas",            f"{n_ent:,}"],
    ["Long. media SUMILLA (words)", f"{long_med:.1f}"],
    ["Objetivo 1M alcanzado",       objetivo],
]

tabla = ax4.table(cellText=metricas[1:], colLabels=metricas[0],
                  cellLoc="left", loc="center", bbox=[0, 0, 1, 1])
tabla.auto_set_font_size(False)
tabla.set_fontsize(10)
for (row, col), cell in tabla.get_celld().items():
    if row == 0:
        cell.set_facecolor("#2c3e50")
        cell.set_text_props(color="white", fontweight="bold")
    elif row % 2 == 0:
        cell.set_facecolor("#ecf0f1")
    cell.set_edgecolor("white")
ax4.set_title("Resumen de Cumplimiento", fontweight="bold")

plt.tight_layout()
plt.savefig("dataset_final_verificacion.png", dpi=150, bbox_inches="tight")
plt.show()
print("Figura guardada: dataset_final_verificacion.png")
```
